#### Lets perform t test independently 

```python
midage=merged[merged['LIFESTAGE']=='MIDAGE SINGLES/COUPLES']
from scipy.stats import ttest_ind

cat1 = midage[midage['PREMIUM_CUSTOMER']=='Mainstream']
cat2 = midage[midage['PREMIUM_CUSTOMER']=='Budget']

ttest_ind(cat1['TOT_SALES'], cat2['TOT_SALES'])
```

```python
midage=merged[merged['LIFESTAGE']=='MIDAGE SINGLES/COUPLES']
from scipy.stats import ttest_ind

cat1 = midage[midage['PREMIUM_CUSTOMER']=='Mainstream']
cat2 = midage[midage['PREMIUM_CUSTOMER']=='Premium']

ttest_ind(cat1['TOT_SALES'], cat2['TOT_SALES'])
```

```python
midage=merged[merged['LIFESTAGE']=='YOUNG SINGLES/COUPLES']
from scipy.stats import ttest_ind

cat1 = midage[midage['PREMIUM_CUSTOMER']=='Mainstream']
cat2 = midage[midage['PREMIUM_CUSTOMER']=='Budget']

ttest_ind(cat1['TOT_SALES'], cat2['TOT_SALES'])
```

```python
midage=merged[merged['LIFESTAGE']=='YOUNG SINGLES/COUPLES']
from scipy.stats import ttest_ind

cat1 = midage[midage['PREMIUM_CUSTOMER']=='Mainstream']
cat2 = midage[midage['PREMIUM_CUSTOMER']=='Premium']

ttest_ind(cat1['TOT_SALES'], cat2['TOT_SALES'])
```

Hence they are statistically significant
